```python
import pandas as pd

df = pd.read_csv('/content/two_categories_dataset.csv',index_col=0)
df.head()
```

# Tareas

Separando textos

```python
#Antes de empezar voy a ver datos del dataset
df.info()
#Mostramos todos los tags que hay
df['tag'].unique()
```

```python
#Separar los textos por categoria
separados = df.groupby('tag')['text'].apply(list)
separados.head()
```

```python
#Para ver si se separamos mostramos todos los textos con tag 0 y con tag 1
print(separados[0][:5])
print(separados[1][:5])
```

Analizando y preprocesando

```python
#Empezamos limpiando el texto
#Hacemos todo minuscula
separadosMIN = separados.apply(lambda x : [i.lower() for i in x])

#Vemos si si estan minusculados
print(separadosMIN[0][:5])
print(separadosMIN[1][:5])
```

```python
import string
#Limpiamos acentos y puntuacion
separadosLIMIN = separadosMIN.apply(lambda x : [i.translate(str.maketrans('', '', string.punctuation)) for i in x])

#Checamos si si se limpio bien
print(separadosLIMIN[0][:5])
print(separadosLIMIN[1][:5])
```

```python
#Ahora quitamos las stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('spanish'))

separadosLIMPIO = separadosLIMIN.apply(lambda x : [i for i in x if i not in stop_words])

#Vemos como quedo
print(separadosLIMPIO[0][:5])
print(separadosLIMPIO[1][:5])
```

```python
#Tokenizamos las palabras
separadosTOKEN = separadosLIMPIO.apply(lambda x : [i.split() for i in x])

#Vemos como quedo
print(separadosTOKEN[0][:5])
print(separadosTOKEN[1][:5])
```

Palabras frecuentes

```python
#K palabras mas frecuentes en cada etiqueta
separadosFREC = separadosTOKEN.apply(lambda x : [nltk.FreqDist(i) for i in x])

#Vemos como quedo
print(separadosFREC[0][:5])
print(separadosFREC[1][:5])
```
